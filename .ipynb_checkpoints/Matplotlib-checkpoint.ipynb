```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
#Study Data Files
mouse_metadata = "Resources/Mouse_metadata.csv"
study_results = "Resources/Study_results.csv"
#Files read
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)
```

```python
#Combine Data
combined_data = pd.merge(mouse_metadata, study_results)
combined_data.head()
```

```python
#Set Index
combined_data = combined_data.set_index("Mouse ID")
```

```python
#Description of the Data
combined_data.describe()
```

```python
#Create summary table
summary_stats = pd.DataFrame(combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].count())
summary_stats
```

```python
#Calculate mean
summary_mean = pd.DataFrame(combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean())
summary_mean
```

```python
#Calculate median
summary_median = pd.DataFrame(combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].median())
summary_median
```

```python
#Calculate variance
summary_variance = pd.DataFrame(combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].var())
summary_variance
```

```python
#Calculate standard deviation
summary_standard_deviation = pd.DataFrame(combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].std())
summary_standard_deviation
```

```python
#Calculate SEM
summary_sem = pd.DataFrame(combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem())
summary_sem
```

```python
#DataFrame
```

```python
#Create DataFrame
data = ({"Mean":summary_mean,
        "Median":summary_median,
        "Variance":summary_variance,
        "Standard Deviation":summary_standard_deviation,
        "Sem":summary_sem})  
index = [0]
summary_stats_df = pd.DataFrame(data, index=index)
summary_stats_df
```

```python
#Bar Graph using Pandas


summary_stats.plot(kind="bar", figsize=(8,5), color = "b", legend=False)

plt.title("Trials per Drug Regiman")

plt.show()
```

```python
#Bar graph using pyploy


#x axis and tick locations
x_axis = np.arange(len(summary_stats))
tick_location = [value for value in x_axis]

plt.figure(figsize = (8,5))
plt.bar(x_axis, summary_stats_df["Trials"], color = "b", width = 1)
plt.xticks(tick_locations, summary_stats_df.index.values, rotation="vertical")

plt.xlim(-0.75, len(x_axis))
plt.ylim(0, max(summary_stats_df["Trials"])+10)

plt.show()
```

```python
#pie plot using pandas

explode = (0.2,0,0,0,0,0,0,0,0,0)

colors = ["red", "blue", "purple", "yellow", "pink", "orange", "green", "lightskyblue", "coral","chartreuse"]


summary_stats.plot(kind="pie", explode=explode, labels=summary_stats.index.values, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140, subplots=True,legend=False)

```
