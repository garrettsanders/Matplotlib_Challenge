```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
```

```python
#Study Data Files
mouse_metadata = "Resources/Mouse_metadata.csv"
study_results = "Resources/Study_results.csv"
#Files read
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)
```

```python
#Combine Data
combined_data = pd.merge(mouse_metadata, study_results)
combined_data.head()
```

```python
#Set Index
combined_data = combined_data.set_index("Mouse ID")
```

```python
#Description of the Data
combined_data.describe()
```

```python
#Create summary table
summary_stats = pd.DataFrame(combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].count())
summary_stats
```

```python
#Calculate mean
summary_mean = pd.DataFrame(combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean())
summary_mean
```

```python
#Calculate median
summary_median = pd.DataFrame(combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].median())
summary_median
```

```python
#Calculate variance
summary_variance = pd.DataFrame(combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].var())
summary_variance
```

```python
#Calculate standard deviation
summary_standard_deviation = pd.DataFrame(combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].std())
summary_standard_deviation
```

```python
#Calculate SEM
summary_sem = pd.DataFrame(combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem())
summary_sem
```

```python
summary_stats_df = pd.DataFrame({"Mean":summary_mean, 
                               "Median":summary_median,
                               "Variance":summary_variance, 
                               "Standard Deviation":summary_standard_deviation,
                               "SEM":summary_sem}, index = "Mouse ID")
summary_stats_df
```
